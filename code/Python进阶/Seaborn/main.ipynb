```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
x = np.arange(8)
y = np.array([1, 5, 4, 3, 2, 3, 1, 6])

df = pd.DataFrame({"x-axis": x, "y-axis": y})

sns.barplot("x-axis", "y-axis", palette="RdBu_r", data=df)
plt.xticks(rotation=90)
plt.show()
```
