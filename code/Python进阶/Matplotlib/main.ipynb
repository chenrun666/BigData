### 绘制x轴和y轴的刻度

```python
x = range(2, 26, 2)
y = [random.randint(15, 30) for i in x]

plt.figure(figsize=(20, 8), dpi=80)

# 设置x轴的刻度
plt.xticks(x)
# plt.yticks(y)
plt.yticks(range(min(y), max(y) + 1))

# 设置x轴的标签
x_ticks_label = [f"{i}:00" for i in x]
plt.xticks(x, x_ticks_label, rotation=45)

plt.plot(x, y)

plt.show()
```

### 设置显示中文

```python
x = range(0, 120)
y = [random.randint(10, 30) for i in range(120)]

plt.figure(figsize=(20, 8), dpi=80)
plt.plot(x, y)

plt.xlabel('时间', rotation=45)

# mac / linux 执行 fc-list
my_font = font_manager.FontProperties(fname='C:\Windows\Fonts\msyh.ttc')
plt.title('每分钟跳动次数', fontproperties=my_font, color='red')

plt.show()
```

### 一图多线

```python
y1 = [1, 0, 1, 1, 2, 4, 4, 5, 6, 5, 4, 3, 1, 1, 1, 6]
y2 = [1, 0, 3, 7, 2, 1, 2, 7, 2, 3, 1, 2, 0, 6, 2, 1]

x = range(11, 27)

plt.figure(figsize=(20, 8), dpi=80)

plt.plot(x, y1, color='red', label='自己')
plt.plot(x, y2, color='blue', label='同事')

# 设置x轴刻度
xtick_labels = [f'{i}岁' for i in x]
plt.xticks(x, xtick_labels, fontproperties=my_font, rotation=45)

plt.grid(alpha=0.4)

plt.legend(prop=my_font, loc='upper right')

plt.show()
```

### 一图多个坐标系子图

```python
x = np.arange(1, 100)

fig = plt.figure(figsize=(20, 10), dpi=80)
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x, x)

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(x, x ** 2)
# 加入网格
ax2.grid(color='red', linestyle='--', linewidth=1, alpha=0.3)

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(x, np.log(x))

plt.show()
```

### 改变坐标轴的默认显示方式

```python
y = range(0, 14, 2)
x = [-3, -2, -1, 0, 1, 2, 3]

# 获得当前图表的图像
ax = plt.gca()

# 设置图形的包围线
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 1))

plt.plot(x, y)

plt.show()
```

## 绘制散点图

```python
"""
3月份每天最高气温
"""
y = [11, 17, 16, 12, 11, 6, 6, 5, 7, 8, 7, 6, 12, 17, 15, 14, 13, 18, 19, 20, 22, 21, 20, 22, 24, 25, 25, 22, 25, 25, 21]
x = range(1, 32)

plt.figure(figsize=(20, 8), dpi=80)

size = np.random.randint(0, 100, 31)  # 散点的大小
# 绘制散点图
plt.scatter(x, y, label='3月份', alpha=0.5, s=size, c='red')

# 设置x轴的刻度
_xticks_labels = [f'3月{i}日' for i in x]
plt.xticks(x[::3], _xticks_labels[::3], fontproperties=my_font, rotation=45)

_yticks_labels = [f'{i}℃' for i in range(min(y), max(y) + 1)]
plt.yticks(range(min(y), max(y) + 1), _yticks_labels)

# 设置标识
plt.xlabel('日期', fontproperties=my_font)
plt.xlabel('温度', fontproperties=my_font)

# 图例
plt.legend(prop=my_font)

plt.show()
```

## 绘制条形度

```python
a = ['流浪地球', '疯狂的外星人', '飞驰人生', '大黄蜂', '熊出没', '新喜剧之王']
b = ['38.14', '19.85', '14.89', '11.36', '6.47', '5.93']

plt.figure(figsize=(20, 8), dpi=80)

rects = plt.bar(a, [float(i) for i in b], width=0.3, color=['r', 'g', 'b', 'r', 'g', 'b'])
plt.xticks(range(len(a)), a, fontproperties=my_font, rotation=45)

plt.yticks(range(0, 41, 5), range(0, 41, 5))

# 条形上增加标注
for rect in rects:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width() / 2, height + 0.5, str(height), ha="center")

plt.show()
```

## 横向条形图

```python
a = ['流浪地球', '疯狂的外星人', '飞驰人生', '大黄蜂', '熊出没', '新喜剧之王']
b = ['38.14', '19.85', '14.89', '11.36', '6.47', '5.93']

plt.figure(figsize=(20, 8), dpi=80)

# rects = plt.barh(range(len(a)), b, height=0.5, color="r")
rects = plt.barh(a, [float(i) for i in b], height=0.5, color=['r', 'g', 'b'] * 2)
plt.yticks(range(len(a)), a, fontproperties=my_font, rotation=45)

for rect in rects:
    width = rect.get_width()
    plt.text(width, rect.get_y() + 0.5 / 2, str(width), va="center")

plt.show()
```

## 直方图

```python
time = [random.randint(100, 200) for _ in range(250)]

plt.figure(figsize=(20, 8), dpi=100)

distance = 2

group_num = int((max(time) - min(time)) / distance)
plt.hist(time, bins=group_num)

plt.xticks(range(min(time), max(time))[::2])

plt.grid(linestyle='--', alpha=0.5)

plt.xlabel("电影时长大小", fontproperties=my_font)
plt.ylabel("电影的数据量", fontproperties=my_font)

plt.show()
```

## 饼状图

```python
label_list = ['第一部分', '第二部分', '第三部分']
size = [55, 35, 10]
color = ['red', 'green', 'blue']
explode = [0, 0.05, 0]

"""
x: (每一块)的比例，如果sum(x) > 1, 会将多出的部分进行均匀；
explode: (每一块)离开中心距离
labels: (每一块)饼图外侧显示的说明文字
labeldistance: 设置标签文本距离圆心位置， 1.1表示1.1倍半径
autopct: 控制饼图内百分比设置
shadow: 设置是否有阴影
strartangle: 起始绘制角度， 默认图是从x轴正方向逆时针画起
pctdistance: 类似于labeldistance, 指定autopct的位置刻度， 默认值为0.6

返回值：
patches: matplotlib.patches.Wedge列表（扇形实例）
l_text: label matplotlib.text.Text列表（标签实例）
p_text: label matplotlib.text.Text列表（百分比标签实例）
"""
plt.figure(figsize=(20, 8), dpi=100)
patches, l_text, p_text = plt.pie(
    size,
    explode=explode,
    colors=color,
    labels=label_list,
    labeldistance=1.1,
    autopct="%1.1f%%",
    shadow=True,
    startangle=90,
    pctdistance=0.6
)

for t in l_text:
    t.set_fontproperties(my_font)

for t in p_text:
    t.set_size(18)

for i in patches:
    i.set_color('pink')
    break

plt.show()
```
