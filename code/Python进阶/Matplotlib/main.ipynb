```python
# 导入模块
import random
import numpy as np 

import matplotlib.pyplot as plt 
from matplotlib import font_manager
```

```python
x = [1, 0, 9]
y = [2, 4, 5]

# 传入x, y轴数据画图
plt.plot(x, y)

# 展示图形
plt.show()
```

## 折线图

### 折线图的颜色和形状设置

```python
x = range(1, 8)
y = [17, 16, 15, 13, 11, 11, 15]

# color
# alpha: 透明度
# linestyle
# linewidth
plt.plot(x, y, color="red", alpha=0.3, linestyle="--", linewidth=3)

plt.show()
```

### 折点样式

```python
x = range(1, 8)
y = [14, 12, 21, 33, 12, 11, 10]

plt.plot(x, y, marker="o", color="red", markersize='20', markeredgecolor='g', markeredgewidth=5)

plt.show()
```

### 设置的图片的大小和保存

```python
x = range(2, 26, 2)
y = [random.randint(15, 30) for i in x]

# 设置图片的大小
# figisize: 指定figure的宽和高， 单位为英寸
# dpi参数指定绘图对象的分辨率， 即每英寸多少像素
plt.figure(figsize=(20, 8), dpi=80)
plt.plot(x, y)

plt.savefig('./t1.png')

plt.show()
```

### 绘制x轴和y轴的刻度

```python
x = range(2, 26, 2)
y = [random.randint(15, 30) for i in x]

plt.figure(figsize=(20, 8), dpi=80)

# 设置x轴的刻度
plt.xticks(x)
# plt.yticks(y)
plt.yticks(range(min(y), max(y) + 1))

# 设置x轴的标签
x_ticks_label = [f"{i}:00" for i in x]
plt.xticks(x, x_ticks_label, rotation=45)

plt.plot(x, y)

plt.show()
```

### 设置显示中文

```python
x = range(0, 120)
y = [random.randint(10, 30) for i in range(120)]

plt.figure(figsize=(20, 8), dpi=80)
plt.plot(x, y)

plt.xlabel('时间', rotation=45)

# mac / linux 执行 fc-list
my_font = font_manager.FontProperties(fname='C:\Windows\Fonts\msyh.ttc')
plt.title('每分钟跳动次数', fontproperties=my_font, color='red')

plt.show()
```

### 一图多线

```python
y1 = [1, 0, 1, 1, 2, 4, 4, 5, 6, 5, 4, 3, 1, 1, 1, 6]
y2 = [1, 0, 3, 7, 2, 1, 2, 7, 2, 3, 1, 2, 0, 6, 2, 1]

x = range(11, 27)

plt.figure(figsize=(20, 8), dpi=80)

plt.plot(x, y1, color='red', label='自己')
plt.plot(x, y2, color='blue', label='同事')

# 设置x轴刻度
xtick_labels = [f'{i}岁' for i in x]
plt.xticks(x, xtick_labels, fontproperties=my_font, rotation=45)

plt.grid(alpha=0.4)

plt.legend(prop=my_font, loc='upper right')

plt.show()
```

### 一图多个坐标系子图

```python
x = np.arange(1, 100)

fig = plt.figure(figsize=(20, 10), dpi=80)
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x, x)

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(x, x ** 2)
# 加入网格
ax2.grid(color='red', linestyle='--', linewidth=1, alpha=0.3)

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(x, np.log(x))

plt.show()
```

### 设置坐标轴范围

```python
x = np.arange(-10, 11, 1)
y = x ** 2

plt.plot(x, y)
plt.ylim(ymin=0)
plt.xlim(xmin=0)

plt.show()
```

### 改变坐标轴的默认显示方式

```python
y = range(0, 14, 2)
x = [-3, -2, -1, 0, 1, 2, 3]

# 获得当前图表的图像
ax = plt.gca()

# 设置图形的包围线
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 1))

plt.plot(x, y)

plt.show()
```
